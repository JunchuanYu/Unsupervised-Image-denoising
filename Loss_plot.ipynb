```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gaussian = [50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300]
gauss_psnr = [25.817, 23.87, 21.25516, 19.47, 17.63594, 16.76, 16.10435, 15.25716, 14.83445, 14.5908, 14.3001]
gauss_loss = [221.9813, 289.13, 527.5016, 715.93, 1174.0908, 1596.9448, 1791.2724, 2042.9526, 2251.355, 2419.4596, 2565.3365]
```

```python
textual = [20, 25, 30, 40, 50, 60, 70, 75]
text_psnr = [24.8188, 22.9147, 21.1505, 19.2265, 18.208, 14.4276, 14.2556, 12.7431]
text_loss = [6.2432, 8.4654, 11.5242, 15.7221, 19.8616, 36.4615, 37.7978, 46.6896]
```

```python
random = [30, 40, 50, 60, 70, 80, 90, 100]
random_psnr = [27.14814, 24.6129, 20.46507, 19.77414, 19.56061, 17.826, 16.813, 11.74212]
random_loss = [3.4774, 3.7518, 3.9055, 4.4887, 4.5543, 3.3009, 5.7496, 10.199]
```

## Gaussian Noise

```python
plt.plot(gaussian, gauss_psnr)
#plt.plot(textual, text_psnr, label='text')
plt.xlim(min(gaussian)-10, max(gaussian)+10)
plt.ylim(0, max(gauss_psnr)+2)
plt.xticks(gaussian)
plt.title("Gaussian noise vs PSNR", fontsize=26)
plt.xlabel("Noise level", fontsize=26)
plt.ylabel("PSNR", fontsize=26)
plt.savefig('gaussian_psnr.png', bbox_inches='tight')
```

```python
plt.plot(gaussian, gauss_loss)
plt.xlim(min(gaussian)-10, max(gaussian)+10)
plt.ylim(0, max(gauss_loss)+200)
plt.xticks(gaussian)
plt.title("Gaussian noise vs Loss", fontsize=26)
plt.xlabel("Noise level", fontsize=26)
plt.ylabel("Loss", fontsize=26)
plt.savefig('gaussian_loss.png', bbox_inches='tight')
```

## Textual Noise

```python
plt.plot(textual, text_psnr, label='gaussian')
plt.xlim(min(textual)-2, max(textual)+2)
plt.ylim(0, max(text_psnr)+2)
plt.title("Textual noise vs PSNR", fontsize=26)
plt.xlabel("Noise level", fontsize=26)
plt.ylabel("PSNR", fontsize=26)
plt.savefig('textual_psnr.png', bbox_inches='tight')
```

```python
plt.plot(textual, text_loss)
plt.xlim(min(textual)-2, max(textual)+2)
plt.ylim(0, max(text_loss)+2)
plt.title("Textual noise vs Loss", fontsize=26)
plt.xlabel("Noise level", fontsize=26)
plt.ylabel("Loss", fontsize=26)
plt.savefig('textual_loss.png', bbox_inches='tight')
```

## Random-impulse Noise

```python
plt.plot(random, random_psnr)
plt.xlim(min(random)-2, max(random)+2)
plt.ylim(0, max(random_psnr)+2)
plt.title("Random noise vs PSNR", fontsize=26)
plt.xlabel("Noise level", fontsize=26)
plt.ylabel("PSNR", fontsize=26)
plt.savefig('random_psnr.png', bbox_inches='tight')
```

```python
plt.plot(random, random_loss)
plt.xlim(min(random)-2, max(random)+2)
plt.ylim(0, max(random_loss)+2)
plt.title("Random noise vs Loss", fontsize=26)
plt.xlabel("Noise level", fontsize=26)
plt.ylabel("Loss", fontsize=26)
plt.savefig('random_loss.png', bbox_inches='tight')
```
